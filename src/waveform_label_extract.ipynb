```python
import pandas as pd
```

```python
labs = pd.read_csv('labs_mon_2023_06_26_fixed_tz.csv')
```

```python
labs.head()
```

```python
labs['CSN'].nunique()
```

```python
def label_rows_extract(label):
    label_rows = labs[labs['Component_name'].str.contains(label, case=False, na=False)]
    print('Number of unique CSNs with label '+label, label_rows['CSN'].nunique())
    print('Levels of label '+label, label_rows['Component_abnormal'].unique())
    return label_rows
```

# ECG

## Troponin

```python
troponin_rows = label_rows_extract('HIGH SENSITIVITY TROPONIN')
```

```python
troponin_rows
```

```python
troponin_rows['Component_abnormal'].unique()
```

```python
abnormality_mapping = {
    'Normal': 0,
    'Abnormal High': 1,
    'Critical High': 2
}

# Assuming 'df' is your DataFrame
troponin_rows['Component_abnormal_label'] = troponin_rows['Component_abnormal'].map(abnormality_mapping)
troponin_rows.head()
```

```python
troponin_rows.to_csv('/home/ubuntu/ED_processed/waveform/consolidated/troponin_rows.csv')
```

```python
trop_labels = troponin_rows.groupby('CSN').apply(assign_trop_label).reset_index(name="Label")
trop_labels
```

```python
def assign_trop_label(group):
    if 'Normal' in group['Component_abnormal'].values:
        return 0  
    elif 'Critical High' in group['Component_abnormal'].values:
        return 1
    else:
        return 2
```

```python
earliest_collection_time = troponin_rows.groupby('CSN')['Collection_time'].min().reset_index()
earliest_collection_time
```

```python
trop_meta_info = pd.merge(trop_labels, earliest_collection_time, on='CSN')
trop_meta_info = trop_meta_info[['CSN', 'Collection_time', 'Label']]
trop_meta_info
```

```python
trop_meta_info.to_csv('/home/ubuntu/ED_processed/waveform/consolidated/trop_meta_info.csv')
```
